```python
import random

class securitysystem:
    def __init__(self):
        self.components = {chr(65+i): random.choice(['Safe','vulnerable']) for i in range(9)}

    def get_components(self):
        return self.components

class securityagent:
    def action(self,security_status):
        for component, status in security_status.items():
            if status == 'vulnerable':
                security_status[component] = 'safe'
                print(f"{component} updated")

def running_agent(agent, develop):
    print("initial state: ",develop.get_components())
    agent.action(develop.get_components())
    print("final state: ",develop.get_components()) 

if __name__ == "__main__":
    develop = securitysystem()
    agent = securityagent()
    running_agent(agent,develop)
```

```python
import random

class Datacenter:
    def __init__(self):
        self.server = {f"server{i+1}":random.choice(['underload','balanced','overload']) for i in range(5)}

    def get_server(self):
        return self.server
    
    def current_state(self):
        print(self.server)

class balancing_agent:
    def action(self,state):
        print("initial server status:", state)

        for component, status in state.items():
            if status == "overload":
                for target, target_status in state.items():
                    if status == "underload":
                        print(f"{component} -> {target}")
                        state[component] = 'balanced'
                        state[target] = 'balanced'
                        break
                         
        print('final server state',state)


def agent_runner(agent, obj):
    server = obj.get_server()
    agent.action(server)

if __name__ == "__main__":
    obj = Datacenter()
    agent = balancing_agent()
    agent_runner(agent,obj)
    
```

```python
import random

class backup:
    def __init__(self):
        self.task = {f"task {i+1}": random.choice(['completed','failed'])for i in range(5)}
    
    def get_task(self):
        return self.task
    
class restoreAgent:
    def action(self,task):
        for task_num, status in task.items():
            if status == "failed":
                task[task_num] = 'completed'
                print(f"{task_num} is restored and completed")

def agent_runner(agent,obj):
    print("initial tasks: ",obj.get_task())
    tasks = obj.get_task()
    agent.action(tasks)
    print("final tasks: ",obj.get_task())
    
if __name__ == "__main__":
    agent = restoreAgent()
    obj = backup()
    agent_runner(agent,obj)
```

```python
import random

class SecuritySystem:
    def __init__(self):
        self.system = {f"component {i+1}": random.choice(["Safe","Low Risk Vulnerability","High Risk Vulnerability"]) for i in range(5)}

    def get_system(self):
        return self.system
    
class Agent_47:
    def action(self,system):
        for component,status in system.items():
            if status == "Low Risk Vulnerability":
                print(f"{component} is vulnerable!")
                print(f"patching {component}....")
                system[component] = 'Safe'
                print(f"{component} successfully patched to Safe")
            if status == "High Risk Vulnerability":
                print(f"{component} is Highly vulnerable warning!")
                print(f"{component} is not patchable")
                
def agent_runner(agent,obj):
    print("initial state:", obj.get_system())
    components = obj.get_system()
    agent.action(components)
    print("final state:", obj.get_system())
    
if __name__ == "__main__":
    env = SecuritySystem()
    agent = Agent_47()
    agent_runner(agent, env)

```

```python
class Hospital:
    def __init__(self):
        self.locations = { 
            101: "room 1",
            102: "room 2",
            103: "room 3",
            104: "room 4",
            105: "room 5",
            106: "storage room",
            107: "Nurse Station"
        }
    def get_locations(self):
        return self.locations
        
class Delivery_agent:
    def action(self,obj):
        for number,room in obj.items():
            if room == "storage room":
                print(f"moving to {room}")
                print(f"Taking medicine")
            if room == "Nurse Station":
                print(f"moving to {room}")
                print(f"Calling Nurse")
            else:        
                print(f"moving to room {room}")
                print(f"scanning patient id in room {room}")
                print(f"delivering medicine to room {room}")

def agent_runner(agent, object):
    location = object.get_locations()
    agent.action(location)

if __name__ == "__main__":
    obj = Hospital()
    robot = Delivery_agent()
    agent_runner(robot, obj)
```

```python
class Structure:
    def __init__(self):
        self.chambers = {
            'a': 'secure', 'b': 'secure', 'c': 'blaze',
            'd': 'secure', 'e': 'blaze', 'f': 'secure',
            'g': 'secure', 'h': 'secure', 'j': 'blaze'
        }

class ExtinguishMachine:
    def execute(self, surroundings, route):
        for chamber in route:
            if surroundings.chambers[chamber] == 'blaze':
                surroundings.chambers[chamber] = 'secure'
                print(f"Fire detected in chamber {chamber} - suppression activated!")
            else:
                print(f"Chamber {chamber} is clear - no action needed")

def deploy_unit(unit, surroundings, route):
    unit.execute(surroundings, route)
    print("Final building status:", surroundings.chambers)

if __name__ == "__main__":
    surroundings = Structure()
    extinguisher = ExtinguishMachine()
    
    inspection_route = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j']
    
    deploy_unit(extinguisher, surroundings, inspection_route)
```
