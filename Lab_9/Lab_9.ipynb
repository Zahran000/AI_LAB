```python
import math

def run_minimax():
    actions = {
        "Stop": [6, 7],
        "Go": [2, 4],
        "Turn": [8, 3]
    }

    best_score = float("-inf")
    best_action = None
    results = {}

    for action in actions:
        worst_case = min(actions[action])
        results[action] = worst_case

        if worst_case > best_score:
            best_score = worst_case
            best_action = action

    return best_score, best_action, results


def run_alpha_beta():
    actions = {
        "Stop": [6, 7],
        "Go": [2, 4],
        "Turn": [8, 3]
    }

    alpha = float("-inf")
    best_score = float("-inf")
    best_action = None
    skipped = []

    for action in actions:

        worst_case = min(actions[action])

        if worst_case <= alpha:
            skipped.append(action)
            continue

        if worst_case > best_score:
            best_score = worst_case
            best_action = action

        alpha = max(alpha, best_score)

    return best_score, best_action, skipped


def compare_methods():
    return {
        "Minimax nodes evaluated": 6,
        "Alpha-Beta nodes evaluated": 5
    }


def explain():
    return "Alpha-Beta pruning avoids exploring branches that cannot improve the final decision."


print("Minimax Result:", run_minimax())
print("Alpha-Beta Result:", run_alpha_beta())
print("Comparison:", compare_methods())
print(explain())
```

```python
import math

game_tree = {
    "A1": [5, 6],
    "A2": [7, 4],
    "D1": [3, 8],
    "D2": [6, 2],
    "G1": [1, 9],
    "G2": [4, 7]
}


def minimax():
    attack = min(max(game_tree["A1"]), max(game_tree["A2"]))
    defend = min(max(game_tree["D1"]), max(game_tree["D2"]))
    gather = min(max(game_tree["G1"]), max(game_tree["G2"]))

    best_value = max(attack, defend, gather)

    if best_value == attack:
        decision = "Attack"
    elif best_value == defend:
        decision = "Defend"
    else:
        decision = "Gather"

    return best_value, decision, {
        "Attack": attack,
        "Defend": defend,
        "Gather": gather
    }


def alpha_beta():
    alpha = -math.inf
    beta = math.inf

    visited = 0
    pruned = []

    def max_step(values, label):
        nonlocal alpha, visited, pruned
        best = -math.inf

        for v in values:
            visited += 1
            best = max(best, v)
            alpha = max(alpha, best)

            if alpha >= beta:
                pruned.append(label)
                break

        return best

    def min_step(values, label):
        nonlocal beta, visited, pruned
        worst = math.inf

        for v in values:
            visited += 1
            worst = min(worst, v)
            beta = min(beta, worst)

            if alpha >= beta:
                pruned.append(label)
                break

        return worst


    A1 = max_step(game_tree["A1"], "A1")
    A2 = max_step(game_tree["A2"], "A2")
    attack = min_step([A1, A2], "Attack")

    D1 = max_step(game_tree["D1"], "D1")
    D2 = max_step(game_tree["D2"], "D2")
    defend = min_step([D1, D2], "Defend")

    G1 = max_step(game_tree["G1"], "G1")
    G2 = max_step(game_tree["G2"], "G2")
    gather = min_step([G1, G2], "Gather")

    root = max(attack, defend, gather)

    return root, visited, pruned


def compare_nodes():
    _, visited, _ = alpha_beta()
    return {
        "Minimax nodes (theoretical)": 12,
        "Alpha-Beta nodes (observed)": visited
    }


print("Minimax:", minimax())
print("Alpha-Beta:", alpha_beta())
print("Comparison:", compare_nodes())
```
