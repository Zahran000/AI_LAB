```python
class SecurityAgent:
    def __init__(self, agent_id, name, status="idle"):
        self.agent_id = agent_id
        self.name = name
        self._status = status 
    
    def get_status(self):
        return self._status
    
    def change_status(self, new_status):
        self._status = new_status

class FirewallAgent(SecurityAgent):
    def traffic_monitor(self):
        self.change_status("active")
        print(f"{self.name}  monitoring network traffic")

class MalwareDetectionAgent(SecurityAgent):
    def file_scanner(self):
        self.change_status("active")
        print(f"{self.name}  scanning for malware")

class AutomationAgent(SecurityAgent):
    def automation_task(self):
        self.change_status("active")
        print(f"{self.name}  running automation")

agent1 = FirewallAgent(101, "firewall ")
agent2 = MalwareDetectionAgent(102, "malware  scanner")
agent3 = AutomationAgent(103, "Auto b ot")

agent1.traffic_monitor()
agent2.file_scanner()
agent3.automation_task()
```

```python

class CyberThreat:
    def __init__(self, threat_id, threat_name, level):
        self.threat_id = threat_id
        self.threat_name = threat_name
        self._level = level
    
    def threat_level(self):
        return self._level
    
    def update_level(self, level):
        self._level = level

class PhishingThreat(CyberThreat):
    def email_analysis(self):
        self.update_level("high")
        print(f"Phishing: {self.threat_name}  email analysis in progress")

class RansomwareThreat(CyberThreat):
    def file_check(self):
        self.update_level("critical")
        print(f"ransomware: {self.threat_name}  file system scanning")

class BotnetThreat(CyberThreat):
    def traffic_analysis(self):
        self.update_level("medium")
        print(f"botnet: {self.threat_name}  network traffic monitoring")

t1 = PhishingThreat("123", "emailscam", "medium")
t2 = RansomwareThreat("123", "fileencryptor", "high")
t3 = BotnetThreat("123", "networkzombie", "low")
t1.email_analysis()
t2.file_check()
t3.traffic_analysis()
```

```python

class ResponseSystem:
    def action(self):
        return "generic response action"

class AlertSystem(ResponseSystem):
    def action(self):
        return "alert notifications dispatched"

class BlockSystem(ResponseSystem):
    def action(self):
        return "malicious activities blocked"

class RecoverSystem(ResponseSystem):
    def action(self):
        return "system recovery initiated"

response_units = [AlertSystem(), BlockSystem(), RecoverSystem()]

for unit in response_units:
    print(unit.action())
```
